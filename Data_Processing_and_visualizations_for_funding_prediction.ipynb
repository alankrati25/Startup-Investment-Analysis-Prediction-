```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define gender categories
def classify_gender(row):
    if row["Male Presenters"] > 0 and row["Female Presenters"] == 0:
        return "Male Only"
    elif row["Female Presenters"] > 0 and row["Male Presenters"] == 0:
        return "Female Only"
    elif row["Male Presenters"] > 0 and row["Female Presenters"] > 0:
        return "Both"
    else:
        return "Unknown"

# Apply classification
df["Gender Category"] = df.apply(classify_gender, axis=1)

# Calculate total startups in each category
total_startups = df["Gender Category"].value_counts()
print(total_startups)

# Calculate startups that got funding
funded_startups = df[df["Accepted Offer"] == "Yes"]["Gender Category"].value_counts()

# Calculate funding success rate (%)
funding_success_rate = (funded_startups / total_startups) * 100
funding_success_rate = funding_success_rate.fillna(0)  # Handle NaN cases



# Calculate non-funded startups
non_funded_startups = total_startups - funded_startups
non_funded_startups = non_funded_startups.fillna(0)  # Handle NaN cases

# Combine data
funding_df = pd.DataFrame({"Funded": funded_startups, "Not Funded": non_funded_startups})
ax=funding_df.plot(kind="bar", figsize=(6,6), color=["green", "red"])
for container in ax.containers:
    ax.bar_label(container, fmt="%d", label_type="center", fontsize=10, color="white", fontweight="bold")
plt.xlabel("Founder Gender Category")
plt.ylabel("Number of Startups")
plt.title("Funded vs. Non-Funded Startups by Gender")
plt.legend(["Funded", "Not Funded"])
plt.show()
```

#Female only starups received highest funding ie ~64%
